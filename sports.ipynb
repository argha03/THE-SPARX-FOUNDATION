```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('/content/matches.csv')
data.head()
```

```python
plt.figure(figsize=(15,10))
sns.countplot(x='season', data=data, palette="winter")  # Specify 'season' as the x-axis
plt.title("Number of matches in each season",fontsize=20)
plt.xlabel("season",fontsize=15)
plt.ylabel("Count",fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(18,10))
sns.countplot(x='winner', data=data, palette="cool")
plt.title("Number of matches won by team",fontsize=20)
plt.xlabel("Team",fontsize=15)
plt.ylabel("No. of wins",fontsize=15)
plt.xticks(rotation=90)
plt.show()
```

```python
data['win_by']=np.where(data['win_by_runs']>0,'Bat first','Bowl first')
```

```python
win=data.win_by.value_counts()
labels=np.array(win.index)
sizes = win.values
colors = ['#ff9999','#66b3ff']
plt.figure(figsize = (10,8))
plt.pie(sizes, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True,startangle=90)
plt.title("Match Result",fontsize=20)
plt.axis('equal')
plt.show()
```

```python
#to plot the no. of matches won by inning wise
plt.figure(figsize = (18,10))
sns.countplot(x='season', hue='win_by', data=data, palette="cool") # Use keyword arguments for x and hue
plt.title("Number of matches won by batting and bowling first",fontsize=20)
plt.xlabel("Season",fontsize=15)
plt.ylabel("Count",fontsize=15)
plt.show()
```

```python
#we will plotpie charton the toss decision
Toss=data.toss_decision.value_counts()
labels=np.array(Toss.index)
sizes = Toss.values
colors = ['#ff9999','#66b3ff']
plt.figure(figsize = (10,8))
plt.pie(sizes, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True,startangle=90)
plt.title("Toss Result",fontsize=20)
plt.axis('equal')
plt.show()
```

```python
plt.figure(figsize = (18,10))
sns.countplot(x='season', hue='toss_decision', data=data, palette="cool")
plt.title("Number of matches won by Toss result",fontsize=20)
plt.xlabel("Season",fontsize=15)
plt.ylabel("Count",fontsize=15)
plt.show()
```

```python
four_data = Data[Data['batsman_runs']==4]
four_data.groupby('batting_team')['batsman_runs'].agg([('runs by fours','sum'),('fours','count')])
```

```python
batsman_four = four_data.groupby('batsman')['batsman_runs'].agg([('four','count')]).reset_index().sort_values('four',ascending=0)
batsman_four.iloc[:10,:].plot('batsman','four',kind='bar',color='black')
plt.title("Numbers of fours hit by playes ",fontsize=20)
plt.xticks(rotation=50)
plt.xlabel("Player name",fontsize=15)
plt.ylabel("No. of fours",fontsize=15)
plt.show()
```

```python
ax=four_data.groupby('batting_team')['batsman_runs'].agg([('four','count')]).reset_index().sort_values('four',ascending=0).plot('batting_team','four',kind='bar',color = 'black')
plt.title("Numbers of fours hit by teams ",fontsize=20)
plt.xticks(rotation=50)
plt.xlabel("Player name",fontsize=15)
plt.ylabel("No. of fours",fontsize=15)
plt.show()
```

```python
six_data = Data[Data['batsman_runs']==6]
six_data.groupby('batting_team')['batsman_runs'].agg([('runs by six','sum'),('sixes','count')])
```

```python
 batsman_six = six_data.groupby('batsman')['batsman_runs'].agg([('six','count')]).reset_index().sort_values('six',ascending=0)
 ax=batsman_six.iloc[:10,:].plot('batsman','six',kind='bar',color = 'orange')
 plt.title("Numbers of sixes hit by playes ",fontsize=20)
 plt.xticks(rotation=50)
 plt.xlabel("Player name",fontsize=15)
 plt.ylabel("No. of sixes",fontsize=15)
 plt.show()
```

```python
# Extract the season and number of sixes
season_sixes = complete_data.groupby('season')['batsman_runs'].agg(lambda x: (x==6).sum())

# Create the bar chart
plt.figure(figsize=(15,10))
sns.barplot(x=season_sixes.index, y=season_sixes.values, palette="winter")

# Add title and labels
plt.title("Number of Sixes Hit in Each Season",fontsize=20)
plt.xlabel("Season",fontsize=15)
plt.ylabel("Number of Sixes",fontsize=15)

# Rotate x-axis labels
plt.xticks(rotation=90)

# Show the plot
plt.show()
```
